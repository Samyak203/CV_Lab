```python
import numpy as np
import skimage
from skimage.io import imread
from skimage import io
from skimage.feature import greycomatrix
from skimage.color import rgb2gray
from skimage import color, feature, img_as_ubyte
import imageio
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('cvjpg.jpg')
io.imshow(image)
io.show
```

```python
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
io.imshow(image)
io.show
```

```python
image7=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
io.imshow(image7)
io.show()
```

```python
co_matrix = skimage.feature.greycomatrix(image7, [5], [0], levels=256, symmetric=True, normed=True)
# Calculate texture features from the co-occurrence matrix
contrast = skimage.feature.greycoprops(co_matrix, 'contrast')
correlation = skimage.feature.greycoprops(co_matrix, 'correlation')
energy = skimage.feature.greycoprops(co_matrix, 'energy')
homogeneity = skimage.feature.greycoprops(co_matrix, 'homogeneity')
# Print the texture features
print("Contrast:", contrast)
print("Correlation:", correlation)
print("Energy:", energy)
print("Homogeneity:", homogeneity)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image7, cmap='gray')
plt.title('Image 1')

plt.subplot(1, 2, 2)
plt.imshow(image3, cmap='gray')
plt.title('Image 2')

plt.tight_layout()
plt.show()

# Display the texture features
print("Image 1 Texture Features:")
print("Contrast:", contrast)
print("Correlation:", correlation)
print("Energy:", energy)
print("Homogeneity:", homogeneity)

print("\nImage 2 Texture Features:")
print("Contrast:", contrast1)
print("Correlation:", correlation1)
print("Energy:", energy1)
print("Homogeneity:", homogeneity1)
```
